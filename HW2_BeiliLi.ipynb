```python
# import necessary package 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

2.(a) Generate 100 random realizations from the random variable X, where X follows the Uniform distribution between 0 and 1

```python
np.random.seed(42)
x=np.random.uniform(0,1, size=100)
x=x.reshape(-1,1)
print(x[:10])
```


(b) For each of the random realization of X above, compute the corresponding 
value of Y that is based on Y = 0.1X +$\mathbf{\varepsilon}$, where $\mathbf{\varepsilon}$ is a random variable that satisfies the normal distribution $N(0,0.01)$.

```python
np.random.seed(42)
error=np.random.normal(0,0.1,100)
error=error.reshape(-1,1)
y=0.1*x+error
print(y[:10])
```


(c) Fit a simple linear regression model for y=0.1X+$\mathbf{\varepsilon}$

```python
lr = LinearRegression()
lr.fit(x,y)
```

```python
y_pred = lr.predict(x)
```

```python
import sklearn.metrics as metrics
print('r2:%.3f'%lr.score(x,y))
print('Coefficient is: %.2f'%lr.coef_)
print('Intercept is :%.2f'%lr.intercept_)
mse=metrics.mean_squared_error(y, y_pred) 
print('MSE:',mse)
```

(d) Generate the graph that includes the scatter plot of the sample, the real line, the fitted line.

```python
plt.xlabel('Predictor - X')
plt.ylabel('Response - Y')
plt.plot(x, y_pred, color='red',label='fitted line')
plt.scatter(x,y,s=5)
plt.plot(x,0.1*x,color='black',label='real line')
plt.legend()
plt.show()
```
